```python
import requests
from bs4 import BeautifulSoup
import csv

my_url = "https://realpython.github.io/fake-jobs/"
# create a csv file
csvfile = open('eggs_sections.csv','w', newline='')
# create a writer to convert data into delimited strings
eggwriter = csv.writer(csvfile)

no_error = True

url = my_url
print("Grabbing...", url)
response = requests.get(url, allow_redirects=False)
print(response.url)
print("Status is", response.status_code)

html = response.text

soup = BeautifulSoup(html, "html.parser")
```

```python
stringval = 'card_content'
class_value = {'class':'card-content'}
```

```python
card_content = soup.find_all('div', attrs=class_value)
```

```python
# title company
for card in range(len(card_content)):
    title = card_content[card].find('h2', attrs={'class':'title is-5'})
    company = card_content[card].find('h3', attrs={'class':'subtitle is-6 company'})
    location = card_content[card].find('p', attrs={'class':'location'})
    # datetime = card_content[card].find('')
    print(title, company)
```

```python

items_soup = soup.find_all('ul', attrs={'class':'book-toc list-unstyled'})
items = items_soup[0]
sections = items.find_all("li", attrs={"class":"book-part"})
chapters = items.find_all("li", attrs={"class":""})

eggwriter.writerow(["Sections"])
for i in range(len(sections)):
    section = sections[i].text
    eggwriter.writerow([section])

csvfile.close()

csvfile = open('eggs_chapters.csv','w', newline='')
# modify writer 
chapters_eggwriter = csv.writer(csvfile)

# Header
chapters_eggwriter.writerow(["Chapters"])
for i in range(len(chapters)):
    chapter = chapters[i].text
    chapters_eggwriter.writerow([chapter])
```
